### To increase image light use increase_light() method
- Attributes :
    - image
    - gamma
    - show

```python
low_light_image = cv.imread("./images/low_light_image.jpg")
tool.increase_light(image=low_light_image, show=True)
print("")
```

### To object detection use object_detection() method
- Attributes :
    - image
    - show
    - save

```python
image = cv.imread("./images/object_detection.jpg")
tool.object_detection(image=image , show=True, save=False)
print("")
```

### To detect geometric shapes use geometric_shapes_detection() method
